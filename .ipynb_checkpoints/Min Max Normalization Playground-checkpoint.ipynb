```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("dataset/diabetes.csv",header=None)
df.head()
```

```python
old_min = np.min(df, axis=0)
old_min
```

```python
old_max = np.max(df, axis=0)
old_max
```

```python
denominator = abs(old_max - old_min)
denominator
```

```python
scaler = np.divide((df - old_min), denominator)
scaler
```

```python
new_max = 1
new_min = 0
new_range = scaler * (new_max - new_min) + new_min
new_range
```

```python
def MinMaxNormalization(data, new_min=0, new_max=1):
    old_min = np.min(data, axis=0)
    old_max = np.max(data, axis=0)
    denominator = abs(old_max - old_min)
    
    return np.divide((data - old_min), denominator) * ( new_max - new_min ) + new_min
```

```python
MinMaxNormalization(df)
```
